Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no compiencen por 28, ya que estamos con códigos postales de Madrid

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dnetro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

```r
#ver qué datos hay en CodigoPostal
unique(airbnb$CodigoPostal)
```

```r
#LIMPIEZA DE DATOS: sustituye - y 28 por NAN
airbnbclean <- airbnb
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "-")] <- NaN
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "28")] <- NaN
```

```r
#LIMPIEZA DE DATOS: pone nan en los que no empiezan por 28 y los vacíos
airbnbclean$CodigoPostal[which(str_sub(airbnbclean$CodigoPostal,1,2) != "28")] <- NaN
```

```r
#LIMPIEZA DE DATOS: cps que contienen \n 
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "28002\n28002")] <- "28002"
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "28051\n28051")] <- "28051"
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "Madrid 28004")] <- "28004"
```

```r
#LIMPIEZA DE DATOS: cps que contienen 4 digitos
airbnbclean[which(str_length(airbnbclean$CodigoPostal) == 4), ] 
```

```r
#LIMPIEZA DE DATOS: cps que contienen 6 digitos
airbnbclean[which(str_length(airbnbclean$CodigoPostal) == 6), ] 
```

```r
#modifico los cps con digitos de mas o menos
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "2805")] <- "28005"
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "2815")] <- "28015"
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "2804")] <- "28004"
airbnbclean$CodigoPostal[which(airbnbclean$CodigoPostal == "280013")] <- "28013"
```

Ayuda: Podeis crear un nuevo dataframe con las columnas "Barrio" y "Freq" que contenga el número de entradas en cada barrio y hacer un merge con el dataframe del punto anterior.

```r
#Creo dataframe con frecuencias por barrio
barrio_freq <- data.frame(summary(airbnb_entire$Barrio))
barrio_freq$Barrio <- row.names(barrio_freq)
row.names(barrio_freq) <- NULL
colnames(barrio_freq) <- c('Frecuency','Barrio')
```

Para únicamente los pisos de alquiler en el barrio de Sol:

Calcular un modelo lineal que combine alguna de estas variables:
* NumBanyos
* NumDormitorios
* MaxOcupantes
* MetrosCuadrados

```r
barrio_sol<-subset(airbnb_entire,Barrio=="Sol")
```

Primero calculamos la correlación para ver como se relacionan estas variables entre sí.

```r
library(GGally)
options(repr.plot.height=10,repr.plot.width=10)
ggpairs(barrio_sol[,c("NumBanyos","NumDormitorios","MaxOcupantes","MetrosCuadrados")],
        lower = list(continuous = wrap("density", alpha = 0.8,size=0.2,color='blue')),
        upper = list(continuous = wrap("cor", size = 8))         
       )
```

Se observa que la correlación entre el número de dormitorios y los metros cuadrados es sorprendentemente baja.
¿Son de fiar esos números?

Mediante un histograma o curvas de densidad podemos descartar números que notienen sentido en el dataframe barrio_sol,
para tener una matriz de correlación que tenga mayor sentido.


```r
#vemos que hay inmuebles con cero habitaciones, no es posible
options(repr.plot.height=5,repr.plot.width=5)
ggplot(data=barrio_sol,aes(x=NumDormitorios))+
        geom_histogram(binwidth = 1, col='black', fill='green', alpha=0.4)
```

```r
#veo que hay metroscuadrados igual a cero
options(repr.plot.height=10,repr.plot.width=10)
ggplot(data=barrio_sol,aes(x=MetrosCuadrados))+
    geom_histogram(binwidth = 5, col='black', fill='green', alpha=0.4)+
    scale_x_continuous(breaks=seq(0,200,length.out=40),labels=round(seq(0,200,length.out=40),))
```

```r
#Elimino los valores sospechosos
barrio_sol <- barrio_sol[barrio_sol$NumDormitorios > 0 ,]
barrio_sol <- barrio_sol[barrio_sol$MetrosCuadrados > 15 ,]
```

Una vez que hayamos filtrado los datos correspondientes calcular el valor o la combinación de valores que mejor nos permite obtener el precio de un inmueble.

```r
library(GGally)
options(repr.plot.height=10,repr.plot.width=10)
ggpairs(barrio_sol[,c("NumBanyos","NumDormitorios","MaxOcupantes","MetrosCuadrados")],
        lower = list(continuous = wrap("density", alpha = 0.8,size=0.2,color='blue')),
        upper = list(continuous = wrap("cor", size = 8))         
       )
```

```r
#modelo de regresión lineal
model<-lm(data=barrio_sol, formula = Precio ~  MaxOcupantes+NumBanyos+NumDormitorios+NumCamas+Puntuacion+MetrosCuadrados)
summary(model)
confint(model)
```

```r
# Parece que las variables que más influyen en el precio son NumDormitorios y Puntuacion, 
# hago otro modelo con esas dos variables
model_2var <-lm(data=barrio_sol, formula = Precio ~  NumDormitorios+Puntuacion)
summary(model_2var)
confint(model_2var)

"Se observa que el R2 mejora. Las variables que mejor predicen el precio son Puntuación y Numero de dormitorios"
```

¿Que variable es más fiable para conocer el precio de un inmueble, el número de habitaciones o los metros cuadrados?

```r
model_hab_m2 <-lm(data=barrio_sol, formula = Precio ~  NumDormitorios+MetrosCuadrados)
summary(model_hab_m2)
confint(model_hab_m2)
"Es más fiable el número de habitaciones"
```

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada metro cuadrado extra del piso?

```r
model_m2 <-lm(data=barrio_sol, formula = Precio ~  MetrosCuadrados)
summary(model_m2)
```

```r
confint(model_m2,level=0.95)
"El precio aumenta 1.0691 euros por cada m2 con un intervalo de confianza del 95% [0.7304517,1.407823]"
"Se puede observar que el valor R2 de este modelo es mucho peor que los anteriores, por tanto predice peor el precio"
```

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada habitación?

```r
model_hab <-lm(data=barrio_sol, formula = Precio ~  NumDormitorios)
summary(model_hab)
```

```r
confint(model_hab,level=0.95)
"El precio aumenta 54.45 euros por cada dormitorio con un intervalo de confianza del 95% [47.34,61.64]"
```

¿Cual es la probabilidad de encontrar, en el barrio de Sol, un apartamento en alquiler con 3 dormitorios?
¿Cual es el intervalo de confianza de esa probabilidad?

```r
prob_3dorm <- nrow(barrio_sol[which(barrio_sol$NumDormitorios == 3) , ]) / nrow(barrio_sol)
paste("La probabilidad de encontrar un dormitorio con 3 habitaciones en Sol es:" , round(prob_3dorm,5)*100 ,"%") 
```

```r
"El intervalo de confianza de esta probabilidad es:"
binom.test(nrow(barrio_sol[which(barrio_sol$NumDormitorios == 3) , ]), nrow(barrio_sol))
```
